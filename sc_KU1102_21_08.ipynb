### Visualisasi Data

#### Visualisasi Listing Date IPO

```python
df['ListingYear'] = pd.DatetimeIndex(df['ListingDate']).year
df1 = df['ListingYear'].value_counts()
df1 = df1.sort_index()
plt.show(df1.plot(kind="barh", xlabel="Banyak Saham IPO", ylabel="Tahun IPO", title ="Frekuensi Tanggal Saham IPO", color="red", figsize=(6,10)))
```

##### Visualisasi Data Total Shares Baru per Tahun Listing

```python
df.groupby("ListingYear")["Shares"].sum().plot(kind="bar",x="ListingYear",y="Shares",color="orange",title="Grafik Jumlah Saham Baru yang Beredar Berdasarkan Tahun Listing")
```

#### Visualisasi Listing Board Saham Indonesia

```python
df["ListingBoard"].value_counts().plot(kind='pie', autopct='%1.1f%%', fontsize=10,title="Persentase Listing Board Saham Indonesia")
```

```python
df.groupby(["Sector","ListingBoard"]).size().unstack().plot(kind="barh", stacked=True, title="Stacked Bar Chart antara ListingBoard dengan Sektor Saham")
```

```python
kosong1={}
df4 = pd.DataFrame(kosong1)
df4["jumlah"]=df.groupby("ListingBoard")["MarketCap"].sum().sort_values()
df4["rerata"]=df.groupby("ListingBoard")["MarketCap"].mean().sort_values()
plt.show(df4.plot(kind="barh", y=["jumlah"], title ="Horizontal Bar Chart Listing Board dengan Jumlah Market Cap"))
plt.show(df4.plot(kind="barh", y=["rerata"], title ="Horizontal Bar Chart Listing Board dengan Rata-Rata Market Cap"))
```

#### Visualisasi Data Sektor vs Kapitalisasi Pasar

```python
kosong={}
df3 = pd.DataFrame(kosong)
df3["jumlah"]=df.groupby("Sector")["MarketCap"].sum().sort_values()
df3["rerata"]=df.groupby("Sector")["MarketCap"].mean().sort_values()
plt.show(df3.plot(kind="barh", y=["jumlah"], title ="Horizontal Bar Chart Sektor Saham dengan Jumlah Market Cap"))
plt.show(df3.plot(kind="barh", y=["rerata"], title ="Horizontal Bar Chart Sektor dengan Rata-Rata Market Cap"))
```

```python
kosong4={}
df4 = pd.DataFrame(kosong4)
df4["Termahal"]=df.groupby("Sector")["LastPrice"].max()
df4["Rerata"]=df.groupby("Sector")["LastPrice"].mean().sort_values()
plt.show(df4.plot(kind="barh", y=["Termahal"], title ="Horizontal Bar Chart Sektor Saham dengan Harga Saham Terakhir Termahal"))
plt.show(df4.plot(kind="barh", y=["Rerata"], title ="Horizontal Bar Chart Sektor Saham dengan Harga Saham Rata-Rata Sektor"))
```

### Correlation Heatmap

```python
corr_df = df[["ListingYear", "Shares", "LastPrice", "MarketCap"]]
pear_corr=corr_df.corr(method='pearson')
corr_arr = np.array(pear_corr)
corr_columns = np.array(corr_df.columns)

fig, ax = plt.subplots(figsize=(8,8))
im = ax.imshow(corr_arr, interpolation='nearest')
fig.colorbar(im, orientation='vertical', fraction = 0.05)

ax.set_xticks(np.arange(len(corr_columns)), labels=corr_columns)
ax.set_yticks(np.arange(len(corr_columns)), labels=corr_columns)

for i in range(len(corr_columns)):
    for j in range(len(corr_columns)):
        text = ax.text(j, i, round(corr_arr[i, j], 2),
                       ha="center", va="center", color="black")

ax.set_title("Correlation Heatmap")
fig.tight_layout()
plt.show()
```
